```python
import os
from collections import defaultdict, Counter
```

```python
def map(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read().lower()
        words = text.split()
        word_tuples = [(word, 1) for word in words]
    return word_tuples
```

```python
def combine(mapped_data):
    combined_data = defaultdict(int)
    for word, count in mapped_data:
        combined_data[word] += count
    return combined_data
```

```python
input_dir = 'data'
```

```python
combined_data_list = []
for filename in os.listdir(input_dir):
    file_path = os.path.join(input_dir, filename)
    mapped_data = map(file_path)
    print(mapped_data)
    combined_data = combine(mapped_data)
    combined_data_list.append(combined_data)
```

```python
combined_data_list[0]
```

```python
def reduce(combined_data_list):
    final_counts = defaultdict(int)
    for combined_data in combined_data_list:
        for word, count in combined_data.items():
            final_counts[word] += count
    return final_counts
```

```python
final_counts = reduce(combined_data_list)
```

```python
top = Counter(final_counts).most_common(20)
```

```python
for word, count in top:
    print(f'{word}: {count}')
```
